```python
"""" Task 1 – Fibonacci (Recursion)"""
```

```python
class Fibonacci:
   

    def __init__(self, n):
        self.n = n

    def fib(self, num):
        
        if num <= 1:
            return num
        return self.fib(num - 1) + self.fib(num - 2)

    def generate(self):
       
        return [self.fib(i) for i in range(self.n)]


# Testing
obj = Fibonacci(10)
print(obj.generate())
```

```python
'''Task 2 – Remove One Item from Set'''
```

```python
class RemoveItemFromSet:
   

    def __init__(self, data_set, item):
        self.data_set = data_set
        self.item = item

    def remove_item(self):
        
        self.data_set.discard(self.item)
        return self.data_set


# Testing
obj = RemoveItemFromSet({1, 2, 3, 4, 5}, 3)
print(obj.remove_item())
```

```python
'''Task 3 – Remove Multiple Items from Set'''
```

```python
class RemoveMultipleItems:
    

    def __init__(self, data_set, items):
        self.data_set = data_set
        self.items = items

    def remove_items(self):
       
        for item in self.items:
            self.data_set.discard(item)
        return self.data_set


# Testing
obj = RemoveMultipleItems({1, 2, 3, 4, 5}, {1, 4})
print(obj.remove_items())
```

```python
'''Task 4 – Find Index in Tuple'''
```

```python
class TupleIndex:
   

    def __init__(self, data_tuple, item):
        self.data_tuple = data_tuple
        self.item = item

    def find_index(self):
        
        return self.data_tuple.index(self.item)


# Testing
obj = TupleIndex(('apple', 'banana', 'orange', 'grape'), 'orange')
print(obj.find_index())
```

```python
'''Task 5 – Sentence to List of Words'''
```

```python
class SentenceSplitter:
  

    def __init__(self, sentence):
        self.sentence = sentence

    def split_sentence(self):
        
        return self.sentence.split()


# Testing
obj = SentenceSplitter("Welcome to Amit")
print(obj.split_sentence())
```

```python
'''Task 6 – Palindrome Check'''
```

```python
class PalindromeChecker:
   

    def __init__(self, text):
        self.text = text

    def is_palindrome(self):
        
        if self.text == self.text[::-1]:
            return "Palindrome"
        return "Not a Palindrome"


# Testing
obj = PalindromeChecker("racecar")
print(obj.is_palindrome())
```

```python
'''Task 7 – Factorial'''
```

```python
class Factorial:
   

    def __init__(self, number):
        self.number = number

    def calculate(self):
        """
        Calculates factorial.
        
        Returns:
        int: Factorial value.
        """
        result = 1
        for i in range(1, self.number + 1):
            result *= i
        return result


# Testing
obj = Factorial(5)
print(obj.calculate())
```

```python
'''Task 8 – Squares from 1 to 10'''
```

```python
class SquaresList:
    

    def __init__(self, n):
        self.n = n

    def generate_squares(self):
        
        return [i ** 2 for i in range(1, self.n + 1)]


# Testing
obj = SquaresList(10)
print(obj.generate_squares())
```

```python
'''Task 9 – Sum of Even Numbers'''
```

```python
class SumEvenNumbers:
    """
    Calculates sum of even numbers in a list.
    
    Attributes:
    numbers (list): List of integers.
    """

    def __init__(self, numbers):
        self.numbers = numbers

    def sum_even(self):
        """
        Calculates sum of even numbers.
        
        Returns:
        int: Sum of even numbers.
        """
        return sum(num for num in self.numbers if num % 2 == 0)


# Testing
obj = SumEvenNumbers([1, 2, 3, 4, 5, 6])
print(obj.sum_even())
```
